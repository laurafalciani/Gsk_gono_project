```python
import pandas as pd
import numpy as np
```

```python
linA_genes = pd.read_csv('lineage_11_10_2021_1536.results.csv')
```

```python
linA_genes
```

```python
filt_linA = linA_genes[linA_genes['Benjamini_H_p'] < 0.01]
```

```python
#filt_linA.to_csv('lineage_scoary_185_Ben01.csv')
```

```python
filt_linA
```

```python
filt_linA['Gene']
```

```python
data = pd.read_csv('gene_oneHotEncoding.csv')
```

```python
lineage = pd.read_csv('lineages.csv',sep='\t', index_col = 0)
lineage = lineage.set_index(data.index)
```

```python
datafilt = data[filt_linA['Gene']].join(lineage)
```

```python
datafilt
```

```python
n = 120
```

```python
from sklearn.model_selection import train_test_split
trainf, testf = train_test_split(datafilt, test_size = 0.2)
X_trainf = trainf.iloc[:,:n-1].to_numpy()
X_testf = testf.iloc[:,:n-1].to_numpy()
y_trainf = trainf['lineage'].to_numpy()
y_testf = testf['lineage'].to_numpy()
```

```python
from sklearn import svm
```

```python
#SVM train con le feature selezionate
linear_svc = svm.SVC(kernel='linear')
linear_svc.fit(X_trainf, y_trainf)
```

```python
from sklearn import metrics
y_predf = linear_svc.predict(X_testf)
metrics.confusion_matrix(y_testf, y_predf)
```

```python
metrics.accuracy_score(y_testf, y_predf)
```

```python
from sklearn import preprocessing
X_trainf_normalized = preprocessing.normalize(X_trainf, norm='l2')

X_testf_normalized = preprocessing.normalize(X_testf, norm='l2')
```

```python
##Logistic Regression
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(penalty ='l2').fit(X_trainf_normalized, y_trainf)
y_predf = clf.predict(X_testf_normalized)
metrics.confusion_matrix(y_testf, y_predf)
```

```python
metrics.accuracy_score(y_testf, y_predf)
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
Xf = datafilt.iloc[:,:-1].to_numpy()
yf = datafilt['lineage'].to_numpy()
X_trainf_normalized = preprocessing.normalize(Xf, norm='l2')
```

```python

scores = cross_val_score(LogisticRegression(penalty ='l2'), X_trainf_normalized, yf, cv=7)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_trainf, y_trainf)
y_predf = rf.predict(X_testf)
```

```python
rf.score(X_testf, y_testf, sample_weight=None)
```

```python
metrics.confusion_matrix(y_testf, y_predf)
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
Xf = datafilt.iloc[:,:-1].to_numpy()
yf = datafilt['lineage'].to_numpy()
X_trainf_normalized = preprocessing.normalize(Xf, norm='l2')
scores = cross_val_score(RandomForestClassifier(), X_trainf_normalized, yf, cv=7)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_trainf, y_trainf)
```

```python
y_predf = neigh.predict(X_testf)
metrics.confusion_matrix(y_testf, y_predf)
```

```python
neigh.score(X_testf, y_testf, sample_weight=None)
```

```python
scores = cross_val_score(KNeighborsClassifier(n_neighbors=3), X_trainf_normalized, yf, cv=7)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```
