```python
lineage = pd.read_csv('lineages.csv',sep='\t', index_col = 0)
lineage = lineage.set_index(data.index)
```

```python
datafilt = data[filt_linA['Gene']].join(lineage)
```

```python
datafilt
```

```python
n = 150
```

```python
from sklearn.model_selection import train_test_split

trainf, testf = train_test_split(datafilt, test_size = 0.2)
X_trainf = trainf.iloc[:,:n].to_numpy()
X_testf = testf.iloc[:,:n].to_numpy()
y_trainf = trainf['lineage'].to_numpy()
y_testf = testf['lineage'].to_numpy()
```

```python
from sklearn import svm
```

```python
#SVM train con le feature selezionate
linear_svc = svm.SVC(kernel='linear')
linear_svc.fit(X_trainf, y_trainf)
```

```python
from sklearn import metrics
y_predf = linear_svc.predict(X_testf)
metrics.confusion_matrix(y_testf, y_predf)
```

```python
metrics.accuracy_score(y_testf, y_predf)
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
Xf = datafilt.iloc[:,:n].to_numpy()
#Xfnorm = preprocessing.normalize(Xf, norm='l2')
yf= datafilt['lineage'].to_numpy()
scores = cross_val_score(linear_svc, Xf, yf, cv=5)
```

```python
scores.mean()
```

```python
from sklearn import preprocessing
X_trainf_normalized = preprocessing.normalize(X_trainf, norm='l2')

X_testf_normalized = preprocessing.normalize(X_testf, norm='l2')
```

```python
##Logistic Regression
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(penalty ='l2').fit(X_trainf_normalized, y_trainf)
y_predf = clf.predict(X_testf_normalized)
metrics.confusion_matrix(y_testf, y_predf)
```

```python
metrics.accuracy_score(y_testf, y_predf)
```
